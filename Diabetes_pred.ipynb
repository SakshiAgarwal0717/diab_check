```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('Downloads/diabetes.csv')
```

```python
X= df.drop(columns = 'Outcome', axis = 1)  #drop- axis = 1 for coln , 0 for row 
Y = df['Outcome']
```

```python
scaler = StandardScaler()
X_sc= scaler.fit_transform(X)
```

```python
X_train , X_test , Y_train , Y_test = train_test_split(X_sc,Y,test_size=0.2, stratify = Y, random_state = 2)
#stratify = yes so that data is distributed in proportion not all types in 0.8 should have diabetes 
```

```python
cl = svm.SVC(kernel='linear')
```

```python
cl.fit(X_train, Y_train)
```

```python
x_pred= cl.predict(X_train)
```

```python
accuracy_score(Y_train,x_pred)*100
```

```python
y_pred = cl.predict(X_test)
```

```python
accuracy_score(Y_test,y_pred)*100
```
